   2. Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы.

**1. В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?**

Принципиальное отличие этих систем между собой в том, что в коллаборативной фильтрации учитываются только взаимодействия юзеров с товарами, а в гибридных рекомендательных системах ещё учитываются и различные внешние признаки этих юзеров и товаров.

**2. Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы.**

Гибридные рекомендательные системы можно использовать в следующих случаях:

- Холодный старт юзера - при появлении нового юзера, когда история взаимодействия с товарами ещё не накоплена.
- Холодный старт товара - при появлении нового товара, когда история взаимодействия с юзерами ещё не накоплена.
